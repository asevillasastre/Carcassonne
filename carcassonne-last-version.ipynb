```python
class Ascii_short():
    def __init__(self, ascii_short):
        self.ascii_short = ascii_short

    def __repr__(self):
        return str(self.ascii_short)
    
    def __print__(self):
        return str(self.ascii_short)
    
    def rotate_90(self):
        self.ascii_short = list(self.ascii_short)
        self.ascii_short[0], self.ascii_short[1], self.ascii_short[2], self.ascii_short[3] = self.ascii_short[3], self.ascii_short[0], self.ascii_short[1], self.ascii_short[2]
        self.ascii_short = "".join(self.ascii_short)
    
    def rotate_180(self):
        self.ascii_short = list(self.ascii_short)
        self.ascii_short[0], self.ascii_short[1], self.ascii_short[2], self.ascii_short[3] = self.ascii_short[2], self.ascii_short[3], self.ascii_short[0], self.ascii_short[1]
        self.ascii_short = "".join(self.ascii_short)

    def rotate_270(self):
        self.ascii_short = list(self.ascii_short)
        self.ascii_short[0], self.ascii_short[1], self.ascii_short[2], self.ascii_short[3] = self.ascii_short[1], self.ascii_short[2], self.ascii_short[3], self.ascii_short[0]
        self.ascii_short = "".join(self.ascii_short)
        
```

```python
class Edges():

    def __init__(self, edges):
        self.edges = edges
    
    def __repr__(self):
        return str(self.edges)
    
    def __print__(self):
        return str(self.edges)
    
    def rotate_90(self):
        up, right, down, left = self.edges["up"], self.edges["right"], self.edges["down"], self.edges["left"]
        self.edges["up"], self.edges["right"], self.edges["down"], self.edges["left"] = left, up, right, down

    def rotate_180(self):
        up, right, down, left = self.edges["up"], self.edges["right"], self.edges["down"], self.edges["left"]
        self.edges["up"], self.edges["right"], self.edges["down"], self.edges["left"] = down, left, up, right

    def rotate_270(self):
        up, right, down, left = self.edges["up"], self.edges["right"], self.edges["down"], self.edges["left"]
        self.edges["up"], self.edges["right"], self.edges["down"], self.edges["left"] = right, down, left, up
        
```

```python
class BoardTile():

    def __init__(self, board, tile):
        self.board = board
        self.tile = tile 

    def __repr__(self):
        min_coord_i, max_cord_i, min_coord_j, max_coord_j = min(self.board.dictionary.keys(), key=lambda x: x[0])[0], max(self.board.dictionary.keys(), key=lambda x: x[0])[0], \
                                                            min(self.board.dictionary.keys(), key=lambda x: x[1])[1], max(self.board.dictionary.keys(), key=lambda x: x[1])[1]
        self.board.matrix = np.full((max_cord_i - min_coord_i + 3, max_coord_j - min_coord_j + 3), None)
        for k in self.board.dictionary.keys():
            x, y = - min_coord_i + k[0] + 1, - min_coord_j + k[1] + 1
            self.board.matrix[x][y] = self.board.dictionary[k].ascii_short
        
        for hole in self.board.valid_holes():
            x, y = - min_coord_i + hole[0] + 1, - min_coord_j + hole[1] + 1
            self.board.matrix[x, y] = None

        character_ord = 65
        for actual_hole in self.valid_holes_at_least_one_tile():
            x, y = - min_coord_i + actual_hole[0] + 1, - min_coord_j + actual_hole[1] + 1
            self.board.matrix[x, y] = chr(character_ord) + "."
            character_ord += 1
        return str(self.board.matrix)

    def valid_holes_tile_rotations(self):
        result = []
        rotations = self.tile.get_rotations()
        # print(rotations)
        holes = self.board.valid_holes()
        for hole in holes:
            degrees = 0
            for rotation in rotations:
                valid = True
                if (hole[0]-1,hole[1]) in self.board.dictionary.keys():
                    if self.board.dictionary[(hole[0]-1,hole[1])].edges.edges["down"] !=  rotation.edges.edges["up"]:
                        valid = False
                if (hole[0]+1,hole[1]) in self.board.dictionary.keys():
                    if self.board.dictionary[(hole[0]+1,hole[1])].edges.edges["up"] !=  rotation.edges.edges["down"]:
                        valid = False
                if (hole[0],hole[1]-1) in self.board.dictionary.keys():
                    if self.board.dictionary[(hole[0],hole[1]-1)].edges.edges["right"] !=  rotation.edges.edges["left"]:
                        valid = False
                if (hole[0],hole[1]+1) in self.board.dictionary.keys():
                    if self.board.dictionary[(hole[0],hole[1]+1)].edges.edges["left"] !=  rotation.edges.edges["right"]:
                        valid = False
                if valid:
                    result.append({"hole":hole,"rotation":rotation,"degrees":degrees})
                degrees += 90
        # print(result)
        return result
    
    def valid_holes_at_least_one_tile(self):
        result = []
        for hole in self.valid_holes_tile_rotations():
            if hole["hole"] not in result:
                result.append(hole["hole"])
        return result
    
```

```python
class Match():
    
    def __init__(self, deck, board, players):
        self.deck = deck
        self.board = board
        self.players = players
    
    def __repr__(self):
        return "Match\n" + str(self.board) + "\n" + str(self.deck)
    
    def __print__(self):
        return "Match\n" + str(self.board) + "\n" + str(self.deck)
    
    def draw_tile(self, tile) -> Tile:
        self.deck.delete_tile(tile)
        return tile

    def draw_random_tile(self):
        total_tiles = sum(self.deck.tiles.values())
        random_num = random.randint(1, total_tiles)
        for tile, times in self.deck.tiles.items():
            random_num -= times
            if random_num <= 0:
                break
        return self.draw_tile(tile)

    def drop_tile(self, tile, coords):
        self.board.drop_tile(tile, coords)
```
